## Google Vision API - Landmark Detection
###### Test Images extracted from Google Images

```
# display images in a directory, landmarks, and lat/long of landmarks

def LandmarkDetection():
  #iterate through a dir and open all files
  for file in glob.glob(directory+'/*'):
    f = open(file, 'rb')
    content = f.read()

    # plot image
    a = plt.imread(f)
    plt.imshow(a)
    plt.show()
    
    #request for Vision API
    image = vision.types.Image(content = content)
    response = client.landmark_detection(image = image)
    landmarks = response.landmark_annotations

    # print landmark and lat/long
    for landmark in landmarks:
      print('Landmark: {}'.format(landmark.description))
      print('Score: {} %'.format(round(landmark.score*100, 2)))
      for location in landmark.locations:
        print('Latitude: {}, Longitude: {} \n'.format(
            round(location.lat_lng.latitude, 5), round(location.lat_lng.longitude, 5)))
        
        
```

```
#define the directory:
directory = r'/content/drive/My Drive/MSDS 462/Landmark Images'

#intialize function
LandmarkDetection()
```
